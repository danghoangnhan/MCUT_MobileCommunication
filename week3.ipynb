```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from package_from_scratch.regression import LinearRegression
```

```python
data = pd.read_csv('./data/data_house.csv').values
data
```

```python
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
model = LinearRegression(lr=0.000001, n_iters=100000)
model.fit(np.hstack((np.ones((N, 1)), x)),y)
```

```python
plt.scatter(x,y)
plt.xlabel('acreage')
plt.ylabel('price')
```

```python
x = np.hstack((np.ones((N, 1)), x))
predict = model.predict(x)
plt.plot((x[0][1], x[N-1][1]),(predict[0], predict[N-1]), 'r')
plt.show()
predict
```
